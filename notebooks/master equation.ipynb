# Master equation

We consider a system made of states $A$, $B$, $C$, ... Each state has a given equilibrium population and the transition rate between any two states can be non-zero and is fixed. The population of any state at any time is stored in the vector $P$. The transition rates are stored in matrix $M$. The evolution of the populations follow the master equation: $\frac{dP}{dt} = MP$.

We implemented this in the `PopulationKinetics` class below.


```python
%matplotlib inline
```

```python
import numpy as np
import logging
from matplotlib import pyplot as plt


class PopulationKinetics:
    """PopulationKinetics
    """
    
    def __init__(self, pop_eq, rates):
        """"""
        self.pop_eq       = pop_eq/np.sum(pop_eq)
        self.nstates      = pop_eq.shape[0]
        self.rates        = rates
        self.ntransitions = np.int(self.nstates*(self.nstates-1)/2)
        self.M            = self.transition_matrix()
        
        try:
            assert self.rates.shape[0] == self.ntransitions, f' for {self.nstates} states, provide {self.ntransitions} rates'
        except AssertionError as error:
            logging.error(error)


    def transition_matrix(self):
        """"""
        M = np.zeros((self.nstates, self.nstates))
        i_rate = 0
        for i in np.arange(self.nstates-1):
            for j in np.arange(i+1,self.nstates):
                M[i,j] = self.rates[i_rate]
                i_rate += 1
        # detailed balance
        for i in np.arange(self.nstates-1):
            for j in np.arange(i+1,self.nstates):
                M[j,i] = M[i,j]*self.pop_eq[j]/self.pop_eq[i]
        # normalize
        for i in np.arange(self.nstates):
            for j in np.arange(self.nstates):
                if(i != j):
                    M[i, i] -= M[j, i]
        return M


    def trajectory(self, pop_init, nsteps=1000, dt=0.01):
        """"""
        self.nsteps   = nsteps
        self.dt       = dt
        self.pop_init = pop_init/np.sum(pop_init)
        self.pop    = self.pop_initialize() 
        self.pop_propagate()
    
        
    def pop_initialize(self):
        try:
            assert self.pop_init.shape[0] == self.pop_eq.shape[0], f' Inconsistency in number of states...'
        except AssertionError as error:
            logging.error(error)
        pop      = np.zeros((self.nsteps, self.nstates))
        pop[0,:] = self.pop_init
        return pop
    
    def pop_propagate(self):
        for i in np.arange(1, self.nsteps):
            self.pop[i,:] = self.pop_update(i-1)
    
    def pop_update(self, i):
        return self.pop[i,:] + self.dt*np.dot(self.M, self.pop[i,:])
    
    def show_pop(self, cumsum=False, labels=None):
        fig = plt.figure(figsize=(4,4), dpi=180)
        plt.xlabel('time')
        plt.ylabel('fraction of population')
        for i in np.arange(self.nstates):
            data = self.pop[:,i]
            if cumsum:
                plt.yscale('log')
                data = np.cumsum(data)
            plt.plot(data, label=f'state {i}')
        if labels is None:
            plt.legend()
        else:
            plt.legend(labels)
        plt.grid()
        plt.show()
```

# SIR model
The SIR model is a 3-state kinetic model with states S (susceptible) and I (infected) rapidly interconverting, and S and/or I slowly transitioning to R (removed). The initial population is 100% susceptible, evolving toward 100% removed.

```python
eps = 1e-3
labels   = ('susceptible','infected','removed')
pop_init = np.array([1. , 0. , 0. ])   
pop_eq   = np.array([eps, eps, 1. ]) 
rates    = np.array([1. , 1e-3, 1e-3 ])     
```

Given these parameters, we build the trajecory below

```python
tseries = PopulationKinetics(pop_eq, rates)
tseries.trajectory(pop_init)
```

and display it now

```python
tseries.show_pop(labels=labels)
```
